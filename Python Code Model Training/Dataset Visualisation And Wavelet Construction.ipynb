```python
#Filterring the dataset - from 1/1/1997 to 15/1/2017
val = df.loc[df['year'] >= 1997]
val = val.loc[val['year'] < 2017]
df = df.loc[ df['year'] == 2017 ]
df = df.loc[ df['month'] == 1 ]
df = df.loc[ df['date'] <= 15 ]
df = pd.concat( [val,df],  ignore_index=True)
df
```

```python
plt.figure(figsize=(15,6))
plt.plot(df['date_in_Fraction'], df["daily total sunspot number"]) #naming year as date_in_Fraction
plt.xlabel("Date ")
plt.xticks([round(i) for i in df['date_in_Fraction']])
plt.ylabel("Daily mean Sunspot Number")
plt.title("Daily Total Sunspot Number");
```

```python
color_list = [
    [[0.7, 0.1, 0.1], '-'],
    [[0.1, 0.7, 0.1], '--'],
    [[0.1, 0.1, 0.7], '-.'],
    [[0.7, 0.7, 0.1], ':'],
    [[0.1, 0.7, 0.7], '-'],
    [[0.7, 0.1, 0.7], '--'],
    [[0.7, 0.7, 0.7], '-.'],
    [[0.3, 0.3, 0.3], ':'],
    [[0.9, 0.5, 0.2], '-'],
    [[0.2, 0.8, 0.4], '--'],
    [[0.6, 0.2, 0.8], '-.'],
    [[0.8, 0.6, 0.2], ':'],
    [[0.2, 0.8, 0.8], '-'],
    [[0.8, 0.2, 0.8], '--'],
    [[0.8, 0.8, 0.8], '-.'],
    [[0.4, 0.4, 0.4], ':'],
    [[0.9, 0.7, 0.4], '-'],
    [[0.5, 0.9, 0.2], '--'],
]
```

```python
#finding rmse and r2 score for reconstructed wavelets for levels 2 - 19 for all  wavelet modes
table = []
min_rmse = [99999,pywt.Modes.modes[0], 0 ]
max_r2 = [0,pywt.Modes.modes[0], 0 ]
for mode in pywt.Modes.modes:
  plt. clf()
  row = [mode]
  for i in range(2,20):
    coeffs = pywt.wavedec(df["daily total sunspot number"], 'db2', level=i,mode=mode)
    reconstructed_signal = pywt.waverec(coeffs, 'db2',mode=mode)
    original_signal = df["daily total sunspot number"]

    rmse = math.sqrt(( (original_signal - reconstructed_signal)**2 ).mean())
    if(rmse < min_rmse[0]):
      min_rmse[0] = rmse
      min_rmse[1] = mode
      min_rmse[0] = i

    #print("RMSE:", rmse)
    #plt.plot(original_signal-reconstructed_signal, color = color_list[i-2][0], linestyle = color_list[i-2][1], label = i)

    r2score=1-(sum((original_signal - reconstructed_signal)**2)/sum(original_signal - reconstructed_signal)**2)

    if(r2score > max_r2[0]):
      max_r2[0] = r2score
      max_r2[1] = mode
      max_r2[2] = i

    #rmse = math.sqrt(( (original_signal - reconstructed_signal)**2 ).mean())
    l = [rmse, r2score]
    row.append(l)
  #plt.legend()
  #plt.xlabel("Time")
  #plt.suptitle("Superimposing both graphs in one")
  table.append(row)
  #plt.show()
headings = [ (str(x) + "th level") for x in range(2,20)]
headings.insert(0, "Mode")
print (tabulate(table, headings))

print("MINIMUM RMSE : ", min_rmse)
print("MAXIMUN R2SCORE : ", max_r2)
```

```python
#january 01, 1997 to january 15, 2017

dd = []
mm = []
yyyy = []
```

```python
def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
```

```python
#adding date column for the dataset
for year in range(1997, 2018):
    feb_days = 29 if is_leap_year(year) else 28
    days_in_month = [31, feb_days, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    for month in range(1, 13):
        for day in range(1, days_in_month[month - 1] + 1):
            if (year == 2017 and month == 1 and day <= 15) or (year < 2017):
                dd.append(day)
                mm.append(month)
                yyyy.append(year)
```

```python
#finding rmse and r2 score for reconstructed wavelets of class c flares for levels 2 - 11 for all  wavelet modes
dictc={"Level":range(2,12)}

rms_min=99999999999999999999
r2_max=-9999999999999999999
rms_final=[]
r2_final=[]

for mode in pywt.Modes.modes:
  temp=[]

  for i in range(2,12):

    coeffs = pywt.wavedec(flare_df['c_class_flares'], 'db2', level=i,mode=mode)
    reconstructed_signal = pywt.waverec(coeffs, 'db2',mode=mode)
    original_signal = flare_df["c_class_flares"]

    rmse = math.sqrt(( (original_signal - reconstructed_signal)**2 ).mean())
    r2=1-(sum((original_signal - reconstructed_signal)**2)/sum(original_signal - reconstructed_signal)**2)

    if rmse<rms_min:
      rms_final=[mode,i]
    if r2>r2_max:
      r2_final=[mode,i]

    temp.append([rmse,r2])
  dictc[mode]=temp
```
